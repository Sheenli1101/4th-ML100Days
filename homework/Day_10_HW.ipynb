```python
import pandas as pd
import numpy as np
#機器學習sklearn套件
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
data_path = 'C:/Users/user/Desktop/'
df_train = pd.read_csv(data_path + 'train.csv')
train_Y = np.log1p(df_train['SalePrice'])    #log1p = log（x+1）
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
df.head()
```

```python
# 顯示 GrLivArea 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['GrLivArea'], y = train_Y)   #train_Y = sales prices
plt.show()

# 做線性迴歸, 觀察分析
#是特徵縮放(Feature scaling)的其中一個方法。特徵縮是是指將數據的各個特徵所依縮放至同一個尺度，以避免某自變量尺度過大而使得其他自變量被忽視，並且有助於模型收斂速度加快。
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv = 5).mean()
```

```python
# 將 GrLivArea 限制在 800 到 2500 以內, 調整離群值
df['GrLivArea'] = df['GrLivArea'].clip(800, 2500)
sns.regplot(x = df['GrLivArea'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業1
#### 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'], y = train_Y)   #train_Y = sales prices
plt.show()
#交叉驗證
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv = 5).mean()
```

```python
#對'1stFlrSF'做上下限的限制以期望達到更高的驗證值
df['1stFlrSF'] = df['1stFlrSF'].clip(800, 1500)
sns.regplot(x = df['1stFlrSF'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
#### 續前題, 去除離群值有兩類方式 : 捨棄離群值(刪除離群的資料) 以及調整離群值,請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
df['1stFlrSF'] = df['1stFlrSF'].clip(800, 1500)
sns.regplot(x = df['1stFlrSF'], y=train_Y)
keep_indexs = (df['1stFlrSF'] > 800) & (df['1stFlrSF'] < 1500)
df = df[keep_indexs]
sns.regplot(x = df['1stFlrSF'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```
